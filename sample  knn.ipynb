```r
data<-iris
data<-data[sample(nrow(data)),]
folds <- cut(seq(1,nrow(iris)),breaks=10,labels=FALSE)
```

```r
folds
```

```r
dist<-function(point1, point2){
    sum=0
    for (i in 1:length(point1)){
        sum=sum+(point2[i]-point1[i])^2
    }
    return(sqrt(sum))
}
```

```r
knn<-function(k, data, loc, target_loc){
    temp<-data
temp$dist=apply(X=temp[,-target_loc],1,FUN=dist, loc)
temp<-temp[order(temp$dist),]
nn=temp[1:k,target_loc]
vote=sort(table(nn),decreasing=TRUE)[1]
val<-names(vote)
return(names(vote)) 
}
```

```r
for (i in 1:nrow(data)){
    pred[i]<-knn(3,data[-i,],data[i,],5)
    print(pred[i])
}
```

```r
data$pred=pred
```

```r
for (k in 20:30)
    {
    score=0
    for(i in 1:10)
        {
        testIndexes <- which(folds==i,arr.ind=TRUE)
        testData <- data[testIndexes, ]
        trainData <- data[-testIndexes, ]
        val<-0
        count<-0
        for (j in 1:nrow(testData))
            {        
            val[j]<-knn(k,trainData,testData[j,],5)
            if (val[j]!=testData[j,5])
                {
                count=count+1
                }
            }
       score=score+(1-count/nrow(testData))
        }
    print(k)
    print(score/10)
    }
```
